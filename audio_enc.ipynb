```python
import os

path = '/home/fathinah.izzati/TJ/filtered/tnj_test/video'
file_names = [i for i in os.listdir(path) ]
out = '/home/fathinah.izzati/TJ/filtered/tnj_test/music/'
if not os.path.exists(out):
    os.mkdir(out)
from coco_mulla.utilities.encodec_utils import extract_rvq
import librosa
import torch
import numpy as np
device='cuda'
sr =32000

for i in file_names:
    i_out = out+i.split('.mp4')[0]+'.npy'
    if os.path.exists(i_out):
        continue
    wav, _ = librosa.load(path+'/'+i , sr=sr, mono=True)
    wav = torch.from_numpy(wav).to(device)[None, None, ...]
    mix_rvq = extract_rvq(wav, sr=sr)
    print(mix_rvq.shape)
    # np.save(i_out, mix_rvq.cpu().numpy())
    # print(i_out)
```
